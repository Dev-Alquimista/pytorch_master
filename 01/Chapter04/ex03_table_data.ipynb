```python
wine_numpy = np.loadtxt("./data/winequality-white.csv", dtype=np.float32, delimiter=";", skiprows=1)
col_list = next(csv.reader(open("./data/winequality-white.csv"), delimiter=";"))
wine_numpy.shape, col_list
```

```python
wine_numpy
```

```python
wineq = torch.from_numpy(wine_numpy)
wineq, wineq.shape, wineq.dtype
```

```python
data = wineq[:, : -1]
data, data.shape
```

```python
target = wineq[:, -1]
target, target.shape
```

```python
target = wineq[:, -1].long()
target, target.shape
target[0]
```

```python
# 4.3.4
# One Hot Encoding 
target_onehot = torch.zeros(target.shape[0], 10)
target_onehot.shape 
target_onehot[0]
```

```python
target_onehot.scatter_(1, target.unsqueeze(1), 1.0)
target_onehot[0]
```

```python
# 4.3.5 
# When to categorize 
data_mean = torch.mean(data, dim=0)
data_mean 
```

```python
data_var = torch.var(data, dim=0)
data_var 
```

```python
data_normalized = (data - data_mean) / torch.sqrt(data_var)
data_normalized.shape, data_normalized
```

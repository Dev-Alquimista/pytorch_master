```python
# 4.1.3 
# Changing the layout 
batch_size= 3 
batch = torch.zeros(batch_size, 3, 256, 256, dtype=torch.uint8)
```

```python
FILE_PATH = "/content/drive/MyDrive/dataset/pytorch/image-cats.zip"
EXTRACTOR_PATH = "/content/data/"
shutil.unpack_archive(FILE_PATH, EXTRACTOR_PATH)
```

```python
filenames = [name for name in os.listdir("./data/") if os.path.splitext(name)[-1] ==".png"]

for i, filename in (enumerate(filenames)):
    img_arr = imageio.imread(os.path.join("./data/", filename))
    
    img_t = torch.from_numpy(img_arr)
    img_t = img_t.permute(2, 0, 1)
    imt_t = img_t[:3]
    batch[i] = img_t

batch.shape
```

```python
# 4.1.4 
# Normalizing the data 
# Neural networks usually work with floating-point tensor as their input 
# Neural networks exhibit the best training performance when the input data ranges roughly from 0 to 1, from -1 to 1 
```

```python
batch = batch.float()
batch /= 255.0
```

```python
n_channels = batch.shape[1]
for c in range(n_channels):
    mean = torch.mean(batch[:, c])
    std = torch.std(batch[:, c])
    batch[:, c] = (batch[:, c] -mean / std)

batch
```
